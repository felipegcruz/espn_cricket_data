```python
import pandas as pd
import math
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
df.columns
```

```python
df.shape
```

```python
df.dtypes
```

```python
df = df.rename(columns={'Mat':'Matches', 'NO':'No_Outs','HS':'Highest_Inns_Score','Ave':'Batting_Avarege','BF':'Balls_Faced','SR':'Batting_Strike_Rate'})
df
```

```python
df.isnull().any()
```

```python
df[df['Balls_Faced'].isna()==1]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].fillna(0)
```

```python
selected_players = ['ED Weekes (WI)', 'CL Walcott (WI)', 'Hon.FS Jackson (ENG)']
result_df = df[df['Player'].isin(selected_players)]
result_df
```

```python
duplicated_rows = df[df.duplicated(keep=False)]
duplicated_rows
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.dtypes
```

```python
df[['Start_Date','End_Date']] = df['Span'].str.split('-',expand=True)
df.head(1)
```

```python
df.rename(columns={'Start_Date':'Start_Year','End_Date':'End_Year'})
df.head(1)
```

```python
df = df.drop(columns='Span')
df.head(1)
```

```python
df.insert(df.columns.get_loc('Player') + 1, 'Start_Date', df.pop('Start_Date'))
df.insert(df.columns.get_loc('Player') + 2, 'End_Date', df.pop('End_Date'))

df.head(1)
```

```python
# Extract 'Name' and 'Country' from 'Player' column
df[['Name', 'Country']] = df['Player'].str.extract(r'(.+)\s\((\w+)\)')

df.head(1)
```

```python
df.insert(df.columns.get_loc('Player') + 1, 'Name', df.pop('Name'))
df.insert(df.columns.get_loc('Player') + 2, 'Country', df.pop('Country'))
df = df.drop(columns='Player',axis=1)
df.head(1)
```

```python
df = df.rename(columns={'Name':'Player'})
df.head()
```

```python
df['Score'] = df['Highest_Inns_Score'].str.extract(r'(\d+)\*').fillna(df['Highest_Inns_Score'])
df = df.drop(columns=['Highest_Inns_Score'])
```

```python
df.head()
```

```python
df = df.rename(columns={'Score': 'Highest_Inns_Score'})
```

```python
df.head()
```

```python
df.insert(df.columns.get_loc('Runs') + 1, 'Highest_Inns_Score', df.pop('Highest_Inns_Score'))
df.head()
```

```python
df['Balls_Faced'] = df['Balls_Faced'].str.extract(r'(\d+)\+').fillna(df['Balls_Faced'])
```

```python
df.head()
```

```python
df[['Start_Date', 'End_Date','Highest_Inns_Score','Balls_Faced']] = df[['Start_Date', 'End_Date','Highest_Inns_Score','Balls_Faced']].astype(int)
df.dtypes
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].apply(lambda x: math.ceil(x))
df
```

```python
df.dtypes
```

```python
pd.set_option('display.max_rows',None)
```

```python
carrer_length = df['Carrer_Length'] = df['End_Date'] - df['Start_Date']
df
```

```python
mean_carrer_lenght = df['Carrer_Length'].mean()
mean_carrer_lenght
```

```python
over_10years_batting_strike_rate = df[df['Carrer_Length'] > 10]['Batting_Strike_Rate']
over_10years_batting_strike_rate
```
